```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
font = {'weight' : 'bold', 'size'   : 30}
plt.rc('font', **font)
colors = sns.color_palette("colorblind")
hatches = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']
from comparative_analysis import ComparativeAnalysis
analysis = ComparativeAnalysis("domainadaptationexperiments.csv")
df = analysis.result_df
df = pd.concat([df[df['training_method'] == 'FedSCDepth(Retrain)'], df[df['training_method'] == 'FedSCDepth']])
df = df.replace({'training_method': 'FedSCDepth(Retrain)'}, {'training_method': 'ERFedSCDeph'})
df = df.replace({'training_method': 'FedSCDepth'}, {'training_method': 'FedSCDeph'})
df_ddad = df[df['test_dataset'] == 'DDAD']
df_kitti = df[df['test_dataset'] == 'KITTI']
legend_configs = dict(loc='best', prop={'weight':'bold'}, title=None)
```

```python
analysis.result_df = df_ddad
analysis.plot_metrics_differences(
    ['test_abs_rel', 'test_sq_rel', 'test_rmse', 'test_rmse_log'], 'full', 
    save_as_pdf=True, colors=colors, hatches=hatches, legend_configs=False
)
analysis.plot_metrics_differences(
    ['test_a1', 'test_a2', 'test_a3'], 'full', 
    save_as_pdf=True, colors=colors, hatches=hatches, legend_configs=legend_configs
)
```

```python
analysis.result_df = df_kitti
analysis.plot_metrics_differences(
    ['test_abs_rel', 'test_sq_rel', 'test_rmse', 'test_rmse_log'], 'full', 
    save_as_pdf=True, colors=colors, hatches=hatches, legend_configs=False
)
analysis.plot_metrics_differences(
    ['test_a1', 'test_a2', 'test_a3'], 'full', 
    save_as_pdf=True, colors=colors, hatches=hatches, legend_configs=legend_configs
)
```
